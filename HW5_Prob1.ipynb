```python
import torch
```

```python
t_c = [0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]

t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
t_u =   t_u / t_u.max()# normalize input
params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
```

```python
def model(t_u, w1, w2, b):
  return w2 * t_u ** 2 + w1 * t_u + b

def loss_fn(t_p, t_c):
     return (t_p - t_c).pow(2).mean()

learning_rate = 0.1
for epoch in range(5000):
  t_p  = model(t_u, *params)
  loss = loss_fn(t_p, t_c)
  loss.backward()

  with torch.no_grad():
      params -= learning_rate * params.grad
      params.grad.zero_()


  if epoch % 500 == 0:
        print(f"Epoch {epoch}: loss={loss.item():.4f}, params={params.data}")
```

```python
## linear model params

linear_params = torch.tensor([43.9613, -17.3046], requires_grad=False)  # [w, b]

# === Define nonlinear model params
nonlinear_params = torch.tensor([ 19.3020,  19.7479, -10.3659], requires_grad=False)

# === Define models ===
def linear_model(t_u, w, b):
    return w * t_u + b

def nonlinear_model(t_u, w2, w1, b):
    return w2 * t_u**2 + w1 * t_u + b

import matplotlib.pyplot as plt

with torch.no_grad():
    t_p_linear = linear_model(t_u, *linear_params)
    t_p_nonlin = nonlinear_model(t_u, *nonlinear_params)

plt.figure(figsize=(7,5))
plt.scatter(t_u, t_c, label='Actual data', color='black')
plt.plot(t_u, t_p_linear, label='Linear model', color='blue')
plt.plot(t_u, t_p_nonlin, label='Nonlinear model', color='red', linestyle='--')
plt.xlabel('Normalized Input Temperature (t_u)')
plt.ylabel('Measured Celsius (t_c)')
plt.title('Linear vs Nonlinear Model Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Linear model loss:", loss_fn(t_p_linear, t_c).item())
print("Nonlinear model loss:", loss_fn(t_p_nonlin, t_c).item())
```
